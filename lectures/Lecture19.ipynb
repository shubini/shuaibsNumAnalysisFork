```julia
P = Legendre()
Q = Normalized(P)
X = jacobimatrix(Q)
n = 10
Xₙ = SymTridiagonal(X[1:n,1:n])
x = eigvals(Xₙ)

g = range(-1,1; length=1000)
plot(g, P[g,n+1]; label="P_$n")
plot!(g, Q[g,n+1]; label="Q_$n")
scatter!(x, zero(x); label="roots")
```

```julia
w = [1/norm(Q[x[j],1:n])^2 for j=1:n] # weights for GQ

# exact for polynomials
sum(x.^(2n-2) .* w)
```

```julia
2/(2n-1)
```

```julia

# High accuracy for things that are smooth
sum(exp.(x) .* w) - (exp(1)-exp(-1))
```
